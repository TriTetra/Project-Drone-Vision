### Step 1: Import Libraries

We import `cv2` for all OpenCV functionality and `numpy` for array manipulation. We also import `os` to handle file paths robustly.

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

### Step 2: Load an Image

We use `cv2.imread()` to load an image from a file. The image is loaded as a NumPy array, which allows for mathematical operations. 

**Important:** All assets are now stored in a central `_assets` directory. We construct a relative path to access them.

```python
# Define the relative path to the assets directory
root = os.getcwd()
imagePath = os.path.join(root,'..' ,'_assets','cat.jpg')

# Check if the image exists before trying to read it
if not os.path.exists(imagePath):
    print(f"Error: Image not found at {imagePath}")
else:
    # Read the image
    image = cv2.imread(imagePath)
    print("Image loaded successfully!")
```

```python
plt.figure("Original Image")
plt.imshow(image)
plt.show()
```

### Step 3: Examine Image Properties

Since the image is a NumPy array, we can easily access its properties.
- **`.shape`**: Returns a tuple of (height, width, channels). Channels represent the color (e.g., 3 for BGR).
- **`.dtype`**: Returns the data type of the pixels (e.g., `uint8` for 0-255).

```python
if 'image' in locals():
    print(f"Image Shape: {image.shape}")
    print(f"Data Type: {image.dtype}")
```

### Step 4: Display an Image

To display the image, we use `cv2.imshow()`. This function opens a window to show the image.

- `cv2.imshow("Window Name", image_variable)`: Creates the display window.
- `cv2.waitKey(0)`: Pauses the script indefinitely until a key is pressed. This is crucial for allowing the image to be displayed.
- `cv2.destroyAllWindows()`: Closes all OpenCV windows.

```python
if 'image' in locals():
    cv2.imshow("Cat", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

### Step 5: Resize an Image

We can change the dimensions of an image using `cv2.resize()`. The new dimensions are passed as a `(width, height)` tuple.

```python
if 'image' in locals():
    new_width = 300
    new_height = 200
    resized_image = cv2.resize(image, (new_width, new_height))
    
    print(f"Original shape: {image.shape}")
    print(f"Resized shape: {resized_image.shape}")
    
    cv2.imshow("Resized Cat", resized_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

### Step 6: Save an Image

Finally, we can save our processed image to a new file using `cv2.imwrite()`. The first argument is the new filename, and the second is the image variable to save.

```python
if 'resized_image' in locals():
    # Save the resized image in the current directory
    save_path = 'cat_resized.jpg'
    cv2.imwrite(save_path, resized_image)
    print(f"Image saved to {os.path.abspath(save_path)}")
```
